```python
# Dropping 3 columns which are not useful in this case.

columns_to_drop=['PassengerId','Name','Ticket','Cabin']
data=data.drop(columns_to_drop,axis=1)
data
```

```python
## Filling the missing values in AGE column.

## Fill missing "Age" values with the mean
data['Age'].fillna(data['Age'].mean(), inplace=True)

data.head()
```

```python
## Doing the normalisation or scaling method.

from sklearn.preprocessing import MinMaxScaler

# Separate the target ("Survived") from the features
target = data['Embarked']
features = data.drop(columns=['Embarked'])

# Identify numerical columns (exclude categorical columns)
numerical_columns = features.select_dtypes(include=['number']).columns

scaler = MinMaxScaler()

# Apply MinMax scaling to the numerical features
features[numerical_columns] = scaler.fit_transform(features[numerical_columns])

# Combine the scaled features and the target back into a DataFrame
data_scaled = pd.concat([features, target], axis=1)

data_scaled.head()
```

```python
## Label Encoding


# Define a mapping for label encoding
sex_mapping = {'male': 0, 'female': 1}

# Apply label encoding to the "Sex" column
data_scaled['Sex'] = data_scaled['Sex'].map(sex_mapping)

# Display the DataFrame after label encoding
print("\nDataFrame after Label Encoding on 'Sex' column:")
data_scaled.head()
```

```python
data_scaled.isnull().sum()
```

```python
## Deleting rows of null values in Embarked column

data_scaled=data_scaled.dropna(subset=['Embarked'])
data_scaled.head()
```

```python
# There are no more null or missing values now.
```
