```python
## To display columns with missing values
columns_with_missing_values = data.columns[data.isna().any()].tolist()

print("Columns with missing values:")
for column in columns_with_missing_values:
    print(column)
```

```python
## Filling the missing values in these 3 columns.

## Fill missing "Age" values with the mean
data['Age'].fillna(data['Age'].mean(), inplace=True)

## Filling the missing value as "Unknown" in Cabin & Embarked column
data['Cabin'].fillna('Unknown', inplace=True)
data['Embarked'].fillna('Unknown', inplace=True)

data
```

```python
## Checking if there are still any missing values

columns_with_missing_values_2 = data.columns[data.isna().any()].tolist()
print("Columns with missing values:")
for column in columns_with_missing_values_2:
    print(column)
else:
    print("No missing values")
```

```python
## Create a box plot for the "Age" column

plt.figure(figsize=(8, 6))
plt.boxplot(data['Age']) 
plt.title('Box Plot for Age')
plt.xlabel('Age')

plt.figure(figsize=(8, 6))
plt.boxplot(data['Fare']) 
plt.title('Box Plot for Fare')
plt.xlabel('Fare')

plt.figure(figsize=(8, 6))
plt.boxplot(data['Pclass']) 
plt.title('Box Plot for Passenger Class')
plt.xlabel('Pclass')

plt.show()
```

```python
## IQR Method

Q1_age = data['Age'].quantile(0.25)
Q3_age = data['Age'].quantile(0.75)
IQR_age = Q3_age - Q1_age
print("IQR of Age column:", IQR_age)

Q1_fare = data['Fare'].quantile(0.25)
Q3_fare = data['Fare'].quantile(0.75)
IQR_fare = Q3_fare - Q1_fare
print("IQR of Fare column:", IQR_fare)

Q1_Pclass = data['Pclass'].quantile(0.25)
Q3_Pclass = data['Pclass'].quantile(0.75)
IQR_Pclass = Q3_Pclass - Q1_Pclass
print("IQR of Pclass column:", IQR_Pclass)
```

```python
data[(data['Age']< Q1_age-1.5* IQR_age) | (data['Age']> Q3_age+1.5* IQR_age)]
data[(data['Fare']< Q1_fare-1.5* IQR_fare) | (data['Fare']> Q3_fare+1.5* IQR_fare)]
data[(data['Pclass']< Q1_Pclass-1.5* IQR_Pclass) | (data['Pclass']> Q3_Pclass+1.5* IQR_Pclass)]
data
```

```python
from sklearn.preprocessing import MinMaxScaler

# Separate the target ("Survived") from the features
target = data['Survived']
features = data.drop(columns=['Survived'])

# Identify numerical columns (exclude categorical columns)
numerical_columns = features.select_dtypes(include=['number']).columns

scaler = MinMaxScaler()

# Apply MinMax scaling to the numerical features
features[numerical_columns] = scaler.fit_transform(features[numerical_columns])

# Combine the scaled features and the target back into a DataFrame
data_scaled = pd.concat([features, target], axis=1)

data_scaled
```
