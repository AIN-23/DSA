Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Histogram Plot

```python
dm=pd.read_csv('/content/diamonds.csv')
```

```python
dm.info()
```

```python
plt.hist(dm['cut'],bins=10,edgecolor='k',color='orange')
plt.xlabel('Cuts')
plt.ylabel('Frequency')
plt.title('Histogram of Cuts')
plt.show()
```

Bar Plot

```python
plt.figure(figsize=(10,8))
sns.barplot(x=dm['cut'], y=dm['price'],hue=dm['clarity'])
plt.show()
```

KDE histogram plot

```python
sns.distplot(a=(dm['price']))
```

Joint Plot

```python
sns.jointplot(x='price',y='carat',data=dm)
```

Heat map

```python
# To load the diamond dataset into a Pandas dataframe
dm_data = sns.load_dataset('diamonds')

# To Compute the correlation matrix of the variables
dm_corr_matrix = dm_data.corr()
print(dm_corr_matrix)

# Creating the heatmap using the `heatmap` function of Seaborn
sns.heatmap(dm_corr_matrix, cmap='coolwarm', annot=True)
```
