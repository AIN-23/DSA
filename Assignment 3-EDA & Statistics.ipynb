```python
## d) What do you think about the scoring variation for math, reading and writing based on
## i. Gender
## ii. Test preparation course
```

```python
## i. Based on gender

plt.figure(figsize=(10, 5))
sns.catplot(x='gender', y='math score', data=df, kind='bar', palette='pastel')
plt.title('Math Scores by Gender')

# Create a categorical plot for reading scores
plt.figure(figsize=(10, 5))
sns.catplot(x='gender', y='reading score', data=df, kind='bar', palette='pastel')
plt.title('Reading Scores by Gender')

# Create a categorical plot for writing scores
plt.figure(figsize=(10, 5))
sns.catplot(x='gender', y='writing score', data=df, kind='bar', palette='pastel')
plt.title('Writing Scores by Gender')

plt.show()
```

```python
## The above graphs are made into a single one using the following code.

# Melt the DataFrame to create a 'score_type' column for different score types (math, reading, writing)
data_melted = pd.melt(df, id_vars=['gender'], value_vars=['math score', 'reading score', 'writing score'], 
                      var_name='score_type', value_name='score')

# Create a grouped bar plot
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x='score_type', y='score', hue='gender', data=data_melted, palette='pastel')
plt.title('Score Variation by Gender and Score Type')
plt.ylabel('Score')
plt.xlabel('Score Type')
plt.legend(title='Gender')
plt.show()


```

 From the above plot we can understand the following points:
1. Male is performing more for maths than female
2. Female is performing more in reading.
3. Female is performing more in reading.

```python
## ii. Based on test preparation course

plt.figure(figsize=(10, 5))
sns.catplot(x='test preparation course', y='math score', data=df, kind='bar', palette='pastel')
plt.title('Math Scores by TPC')

# Create a categorical plot for reading scores
plt.figure(figsize=(10, 5))
sns.catplot(x='test preparation course', y='reading score', data=df, kind='bar', palette='pastel')
plt.title('Reading Scores by TPC')

# Create a categorical plot for writing scores
plt.figure(figsize=(10, 5))
sns.catplot(x='test preparation course', y='writing score', data=df, kind='bar', palette='pastel')
plt.title('Writing Scores by TPC')

plt.show()
```

```python
## The above graphs are made into a single one using the following code.

# Melt the DataFrame to create a 'score_type' column for different score types (math, reading, writing)
data_melted = pd.melt(df, id_vars=['test preparation course'], value_vars=['math score', 'reading score', 'writing score'], 
                      var_name='score_type', value_name='score')

# Create a grouped bar plot
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x='score_type', y='score', hue='test preparation course', data=data_melted, palette='pastel')
plt.title('Score Variation by TPC and Score Type')
plt.ylabel('Score')
plt.xlabel('Score Type')
plt.legend(title='test preparation course')
plt.show()
 
```

From the above plot, it's clear that the ones who have completed the test course are performing well in maths, reading and writing.

```python
## Question 2

##  A company started to invest in digital marketing as a new way of their product promotions. For that they collected data and decided to carry out a study on it.
```

```python
import pandas as pd
import scipy.stats as stats
```

```python
data=pd.read_csv(r"C:\Users\Workstation X\Desktop\ICT\Assignment 3\Sales_add.csv")
data
```

```python
data.info()
```

```python
## a) The company wishes to clarify whether there is any increase in sales after stepping into digital marketing.



# Perform a paired t-test
t_statistic, p_value = stats.ttest_rel(data['Sales_before_digital_add(in $)'], data['Sales_After_digital_add(in $)'])

# Define the significance level (alpha)
alpha = 0.05

# Check if the p-value is less than alpha to determine statistical significance
if p_value < alpha:
    print("There is a statistically significant increase in sales after digital marketing.")
else:
    print("There is no statistically significant increase in sales after digital marketing.")
```

```python
## b) The company needs to check whether there is any dependency between the features “Region” and “Manager”.



from scipy.stats import chi2_contingency

# Create a contingency table (cross-tabulation)
contingency_table = pd.crosstab(data['Region'], data['Manager'])

# Perform the Chi-squared test for independence
chi2_stat, p_value,dof,exp = chi2_contingency(contingency_table)

# Define the significance level (alpha)
alpha = 0.05

# Check if the p-value is less than alpha to determine statistical significance
if p_value < alpha:
    print("There is a statistically significant dependency between Region and Manager.")
else:
    print("There is no statistically significant dependency between Region and Manager.")

```
